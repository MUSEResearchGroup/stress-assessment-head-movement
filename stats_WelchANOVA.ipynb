```python
import scipy.io
import numpy as np
#Import data from MATLAB
abs_data = scipy.io.loadmat('Stats-Stroop STFT/absData.mat')
phase_data = scipy.io.loadmat('Stats-Stroop STFT/phaseData.mat')

Observations_disp_abs = abs_data['Observations_disp_abs'].squeeze()
Observations_speed_abs = abs_data['Observations_speed_abs'].squeeze()
Observations_xy_abs = abs_data['Observations_xy_abs'].squeeze()
Observations_xz_abs = abs_data['Observations_xz_abs'].squeeze()
Observations_yz_abs = abs_data['Observations_yz_abs'].squeeze()

Observations_disp_phase = phase_data['Observations_disp_phase'].squeeze()
Observations_speed_phase = phase_data['Observations_speed_phase'].squeeze()
Observations_xy_phase = phase_data['Observations_xy_phase'].squeeze()
Observations_xz_phase = phase_data['Observations_xz_phase'].squeeze()
Observations_yz_phase = phase_data['Observations_yz_phase'].squeeze()

grp_disp_abs = abs_data['grp_disp_abs'].squeeze()
grp_speed_abs = abs_data['grp_speed_abs'].squeeze()
grp_xy_abs = abs_data['grp_xy_abs'].squeeze()
grp_xz_abs = abs_data['grp_xz_abs'].squeeze()
grp_yz_abs = abs_data['grp_yz_abs'].squeeze()

grp_disp_phase = phase_data['grp_disp_phase'].squeeze()
grp_speed_phase = phase_data['grp_speed_phase'].squeeze()
grp_xy_phase = phase_data['grp_xy_phase'].squeeze()
grp_xz_phase = phase_data['grp_xz_phase'].squeeze()
grp_yz_phase = phase_data['grp_yz_phase'].squeeze()

Participants_disp_abs = abs_data['Participants_disp_abs'].squeeze()
Participants_speed_abs = abs_data['Participants_speed_abs'].squeeze()
Participants_xy_abs = abs_data['Participants_xy_abs'].squeeze()
Participants_xz_abs = abs_data['Participants_xz_abs'].squeeze()
Participants_yz_abs = abs_data['Participants_yz_abs'].squeeze()

Participants_disp_phase = phase_data['Participants_disp_phase'].squeeze()
Participants_speed_phase = phase_data['Participants_speed_phase'].squeeze()
Participants_xy_phase = phase_data['Participants_xy_phase'].squeeze()
Participants_xz_phase = phase_data['Participants_yz_phase'].squeeze()
Participants_yz_phase = phase_data['Participants_xz_phase'].squeeze()
```

```python
#Welch's ANOVA: spostamento abs

import pingouin as pg
import pandas as pd

#create DataFrame
df = pd.DataFrame({'Abs_disp': Observations_disp_abs,
                   'phase': grp_disp_abs}) 

#perform Welch's ANOVA
pg.welch_anova(dv='Abs_disp', between='phase', data=df)

# Post-hoc Games-Howell

pg.pairwise_gameshowell(dv='Abs_disp', between='phase', data=df)
```
